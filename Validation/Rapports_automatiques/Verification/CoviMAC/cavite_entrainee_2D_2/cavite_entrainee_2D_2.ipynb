## Comparaisons a une solution de reference
La reference est dans le src
### Comparaison
Reference :  The lid-driven square cavity flow:  numerical solution with a 1024 x 1024 grid, Marchi et al.,Journal of the Brazilian Society of Mechanical Sciences and Engineering, 2009

```python
from trustutils.jupyter import plot 
import numpy as np
import matplotlib.pyplot as plt 
from trustutils import files as tf 

a=plot.Graph("Comparaison")  
data=plot.loadText("./ref_vx.txt")
a.add(data[0],data[1],marker="*",label="vx, reference")
data=plot.loadText("./ref_vy.txt")
a.add(data[0],data[1],marker="*",label="vy, reference")  

a.addSegment("./cartesian/jdd_VX.son",label="vx, cartesian mesh, PolyMAC") 
a.addSegment("./cartesian/jdd_VY.son",var="y",label="vy, cartesian mesh, PolyMAC")

a.addSegment("./triangle/jdd_VX.son",label="vx, triangle mesh, PolyMAC") 
a.addSegment("./triangle/jdd_VY.son",var="y",label="vy, triangle mesh, PolyMAC")

a.addSegment("./non_conformal/jdd_VX.son",label="vx, non conformal mesh, PolyMAC") 
a.addSegment("./non_conformal/jdd_VY.son",var="y",label="vy, non conformal mesh, PolyMAC")

a.label('position','velocity')
a.visu()
```
